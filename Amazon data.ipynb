```python
#Top 10 most revenue earned by region
fig1,ax1=plt.subplots(figsize=(12,7))
dataset.groupby('Region')['Total Revenue'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

```python
#Top 10 most revenue earned by country
fig1,ax1=plt.subplots(figsize=(12,7))
dataset.groupby('Country')['Total Revenue'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
